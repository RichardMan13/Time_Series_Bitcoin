### ESTRATEGIA 1 BUY IF PREDICT > ACTUAL_PRICE AND CONTRARIWISE FOR SELLING

```python
data_resume = {
    'model': [],
    'periodicity': [],
    'buy_profit': [],
    'sell_profit': [],
    'buy_loses': [],
    'sell_loses': [],
    'sell_wins': [],
    'buy_wins': [],
    'sell_misses': [],
    'buy_misses': [],
    'buy_win_ratio': [],
    'sell_win_ratio': [],
    'total_win_ratio': [],
    'buy_balance': [],
    'sell_balance': [],
    'total_balance': [],
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1-score': [],
}

data_detailed = {
    'date_test': [],
    'model': [],
    'periodicity': [],
    'expected_balance': [],
    'real_balance': [],
}

min_pct_for_operation = 0.00
usd_amount = 100
list_files = os.listdir('datasets/train_test_remake')
# list_files = ['test_lstm_1d_50_epochs.csv','test_lstm_1d_100_epochs.csv', 'test_lstm_1d_150_epochs.csv', 'test_lstm_1d_200_epochs.csv', 
#               'test_lstm_1d_250_epochs.csv', 'test_lstm_1d_300_epochs.csv', 'test_lstm_1d_350_epochs.csv', 'test_lstm_1d_400_epochs.csv'] #

for file in list_files:
    
    file_aux = file.split('_')
    
    if file_aux[0] == 'test':
                        
        model = file_aux[1].upper()
        periodicity = file_aux[2].split('.')[0].upper()
        
        if model == 'NAIVE':
            continue
#         data_resume['model'].append(model+'_'+file_aux[3])
        data_resume['model'].append(model)
        data_resume['periodicity'].append(periodicity)
        
        df_aux = pd.read_csv('datasets/train_test_remake/'+file, sep = '\t')
#         df_aux = pd.read_csv('/home/ricardo/Documentos/TCC_FILES/TCC/'+file, sep = '\t')      
        
        
        buy_balance = 0
        sell_balance = 0
    
        buy_win_ratio = 0
        sell_win_ratio = 0
        
        buy_profit = 0
        sell_profit = 0
        buy_loses = 0
        sell_loses = 0
        
        buy_wins = 0
        sell_wins = 0
        buy_misses = 0
        sell_misses = 0
        
        accumulated_expected_balance = 0
        type_operation = None
        
        for i in range(len(df_aux) -1):
            print('---')
            y_test_predict = df_aux.iloc[i+1]['y_test_predict']
            y_test_real = df_aux.iloc[i+1]['y_test']
            y_test = df_aux.iloc[i]['y_test']
            date_test = df_aux.iloc[i]['date_test']
            
#             print('y_test_predict->', y_test_predict)
#             print('y_test_real->', y_test_real)
#             print('y_test->', y_test)
    
            if y_test_predict > y_test:
                if (y_test_predict/y_test - 1)*100 >= min_pct_for_operation:
                    print("BUY")
                    type_operation = 'BUY'
                    expected_balance = (y_test_predict/y_test - 1)*usd_amount
                    real_balance = (y_test_real/y_test - 1)*usd_amount
            
            else:
                if (y_test/y_test_predict - 1)*100 >= min_pct_for_operation:
                    print("SELL")
                    type_operation = 'SELL'
                    expected_balance = (y_test/y_test_predict - 1)*usd_amount
                    real_balance = (y_test/y_test_real - 1)*usd_amount
                
            if type_operation == 'BUY':
                if real_balance > 0:
                    buy_profit += real_balance
                    buy_wins += 1

                else:
                    buy_loses += real_balance
                    buy_misses += 1

                accumulated_expected_balance += expected_balance

            if type_operation == 'SELL':
                if real_balance > 0:
                    sell_profit += real_balance
                    sell_wins += 1

                else:
                    sell_loses += real_balance
                    sell_misses += 1

                accumulated_expected_balance += expected_balance

#                 print('expected_balance', expected_balance)
#                 print('real_balance', real_balance)

            type_operation = None

            print('---')

            data_detailed['date_test'].append(date_test)
            data_detailed['model'].append(model)
            data_detailed['periodicity'].append(periodicity)
            data_detailed['expected_balance'].append(accumulated_expected_balance)
            data_detailed['real_balance'].append(buy_profit + sell_profit + buy_loses + sell_loses)
        
        buy_win_ratio = round(100*buy_wins/(buy_wins+buy_misses), 2)
        sell_win_ratio = round(100*sell_wins/(sell_wins+sell_misses), 2)
        wins = buy_wins + sell_wins
        misses = buy_misses + sell_misses
        total_win_ratio = round(100*wins/(wins+misses), 2)
        
        buy_balance = round(buy_profit + buy_loses, 2)
        sell_balance = round(sell_profit + sell_loses, 2)
        total_balance = round(buy_balance + sell_balance, 2)
        
        accuracy = (buy_wins+sell_misses)/(buy_wins+buy_misses+sell_misses+sell_wins)
        precision = buy_wins/(buy_wins+buy_misses)
        recall = buy_wins/(buy_wins+sell_misses)
        f1_score = 2*precision*recall/(precision+recall)
        
        data_resume['buy_profit'].append(buy_profit)
        data_resume['sell_profit'].append(sell_profit)
        data_resume['buy_loses'].append(buy_loses)
        data_resume['sell_loses'].append(sell_loses)
        data_resume['buy_wins'].append(buy_wins)
        data_resume['sell_wins'].append(sell_wins)
        data_resume['buy_misses'].append(buy_misses)
        data_resume['sell_misses'].append(sell_misses)
        data_resume['buy_win_ratio'].append(buy_win_ratio)
        data_resume['sell_win_ratio'].append(sell_win_ratio)
        data_resume['total_win_ratio'].append(total_win_ratio)
        data_resume['buy_balance'].append(buy_balance)
        data_resume['sell_balance'].append(sell_balance)
        data_resume['total_balance'].append(total_balance)
        data_resume['accuracy'].append(accuracy)
        data_resume['precision'].append(precision)
        data_resume['recall'].append(recall)
        data_resume['f1-score'].append(f1_score)
                
        
df_detailed_strategy_1 = pd.DataFrame(data_detailed).sort_values(by = 'date_test').reset_index(drop = True)
df_resume_strategy_1 = pd.DataFrame(data_resume).sort_values(by = 'model').reset_index(drop = True)
```

### ESTRATEGIA 2 BUY AND HOLD

```python
data_resume = {
    'periodicity': [],
    'profit': [],
    'loses': [],
    'wins': [],
    'misses': [],
    'win_ratio': [],
    'balance': [],
    # 'accuracy': [],
    # 'precision': [],
    # 'recall': [],
    # 'f1-score': [],    
}

data_detailed = {
    'date_test': [],
    'real_balance': [],
    'periodicity': [],
}

min_pct_for_operation = 0.00
usd_amount = 100
        
list_df = [df_gemini_price_d, df_gemini_price_1h]

for i in range(len(list_df)):
    
    df = list_df[i]
    train_size = int(len(df) * 0.80)
    df_aux = df[train_size:]
    
    if i == 0:
        periodicity = '1D'
    else:
        periodicity = '1H'
        
    balance = 0
    win_ratio = 0

    profit = 0
    wins = 0
    loses = 0
    misses = 0

    is_operation = None

    for i in range(len(df_aux) -1):
        print('---')
        y_test_real = df_aux.iloc[i+1]['Close']
        y_test = df_aux.iloc[i]['Close']
        date_test = df_aux.iloc[i]['Date']

    #             print('y_test_predict->', y_test_predict)
    #             print('y_test_real->', y_test_real)
    #             print('y_test->', y_test)



        print("BUY")
        is_operation = True
        real_balance = (y_test_real/y_test - 1)*usd_amount

        if is_operation == True:
            if real_balance > 0:
                profit += real_balance
                wins += 1

            else:
                loses += real_balance
                misses += 1

    #                 print('expected_balance', expected_balance)
        print('real_balance', real_balance)

        is_operation = False

        print('---')

        data_detailed['date_test'].append(date_test)
        data_detailed['periodicity'].append(periodicity)
        data_detailed['real_balance'].append(profit + loses)

    win_ratio = round(100*wins/(wins+misses), 2)
    balance = round(profit + loses, 2)
    
    # accuracy = (buy_wins+sell_misses)/(buy_wins+buy_misses+sell_misses+sell_wins)
    # precision = buy_wins/(buy_wins+buy_misses)
    # recall = buy_wins/(buy_wins+sell_misses)
    # f1_score = 2*precision*recall/(precision+recall)

    data_resume['periodicity'].append(periodicity)
    data_resume['profit'].append(profit)
    data_resume['loses'].append(loses)
    data_resume['wins'].append(wins)
    data_resume['misses'].append(misses)
    data_resume['win_ratio'].append(win_ratio)
    data_resume['balance'].append(balance)
    # data_resume['accuracy'].append(accuracy)
    # data_resume['precision'].append(precision)
    # data_resume['recall'].append(recall)
    # data_resume['f1-score'].append(f1_score)

df_detailed_strategy_2 = pd.DataFrame(data_detailed).sort_values(by = 'date_test').reset_index(drop = True)
df_resume_strategy_2 = pd.DataFrame(data_resume).sort_values(by = 'periodicity').reset_index(drop = True)
```

### ESTRATEGIA 3 VOTING SYSTEM

```python
data_resume = {
    'periodicity': [],
    'buy_profit': [],
    'sell_profit': [],
    'buy_loses': [],
    'sell_loses': [],
    'sell_wins': [],
    'buy_wins': [],
    'sell_misses': [],
    'buy_misses': [],
    'buy_win_ratio': [],
    'sell_win_ratio': [],
    'total_win_ratio': [],
    'buy_balance': [],
    'sell_balance': [],
    'total_balance': [],
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1-score': [],
}

data_detailed_aux = {
    'date_test': [],
    'model': [],
    'periodicity': [],
    'operation_type': [],
    'real_balance': [],
}

min_pct_for_operation = 0.00
usd_amount = 100
list_files = os.listdir('datasets/train_test_remake')

print("starting making the individuals predictions")
for file in list_files:
    
    file_aux = file.split('_')

    if file_aux[0] == 'test':

        model = file_aux[1].upper()
        periodicity = file_aux[2].split('.')[0].upper()
        
        if model == 'NAIVE':
            continue
        
        df_aux = pd.read_csv('datasets/train_test_remake/'+file, sep = '\t')      
                
        for i in range(len(df_aux) -1):
            
            y_test_predict = df_aux.iloc[i+1]['y_test_predict']
            y_test_real = df_aux.iloc[i+1]['y_test']
            y_test = df_aux.iloc[i]['y_test']
            date_test = df_aux.iloc[i]['date_test']
            
#             print('y_test_predict->', y_test_predict)
#             print('y_test_real->', y_test_real)
#             print('y_test->', y_test)
    
            if y_test_predict > y_test:
                if (y_test_predict/y_test - 1)*100 >= min_pct_for_operation:
                    operation = 1
                    real_balance = (y_test_real/y_test - 1)*usd_amount
            
            else:
                if (y_test/y_test_predict - 1)*100 >= min_pct_for_operation:
                    operation = -1
                    real_balance = (y_test/y_test_real - 1)*usd_amount
            
            data_detailed_aux['date_test'].append(date_test)
            data_detailed_aux['model'].append(model)
            data_detailed_aux['periodicity'].append(periodicity)
            data_detailed_aux['operation_type'].append(operation)
            data_detailed_aux['real_balance'].append(real_balance)
            
print("finish")
print("starting voting system")

df_aux = pd.DataFrame(data_detailed_aux).sort_values(by = 'date_test').reset_index(drop = True)

df_detailed_strategy_3 = pd.DataFrame([])

list_periodicity = df_aux['periodicity'].unique()

for periodicity in list_periodicity:
    
#     balance = 0
#     win_ratio = 0

#     profit = 0
#     wins = 0
#     loses = 0
#     misses = 0
    
    buy_balance = 0
    sell_balance = 0

    buy_win_ratio = 0
    sell_win_ratio = 0

    buy_profit = 0
    sell_profit = 0
    buy_loses = 0
    sell_loses = 0

    buy_wins = 0
    sell_wins = 0
    buy_misses = 0
    sell_misses = 0

    accumulated_expected_balance = 0
    is_operation = None
        
    df_filtered_by_periodicity = df_aux[ df_aux['periodicity'] == periodicity ]
    
    list_date_test = df_filtered_by_periodicity['date_test'].unique()
    
    for date_test in list_date_test:
        
        mask_date_test = df_filtered_by_periodicity['date_test'] == date_test
        df_filtered_by_date_test = df_filtered_by_periodicity[mask_date_test]
        
        mask_operation_type = df_filtered_by_date_test['operation_type'] > 0
        # print('---')
        
        if df_filtered_by_date_test[mask_operation_type]['operation_type'].sum() > 3:
            
#             print("BUY")
            is_operation = True
            voting = df_filtered_by_date_test[mask_operation_type]['operation_type'].sum()
            operation_type = 1
            # print('voting', voting)

        elif df_filtered_by_date_test[~mask_operation_type]['operation_type'].sum() < -3:
            
#             print("SELL")
            is_operation = True
            voting = df_filtered_by_date_test[~mask_operation_type]['operation_type'].sum()
            operation_type = -1
            # print('voting', voting)

        else:
#             print('DO NOTHING')
            voting = 0
            operation_type = 0

        df_row = df_filtered_by_date_test.drop(columns=['model']).iloc[0]
        df_row['voting'] = voting
        df_row['operation_type'] = operation_type
        real_balance = df_row['real_balance']

        if operation_type == 1:
            if real_balance > 0:
                buy_profit += real_balance
                buy_wins += 1

            else:
                buy_loses += real_balance
                buy_misses += 1
                
            balance = buy_profit + buy_loses
            
        if operation_type == -1:
            if real_balance > 0:
                sell_profit += real_balance
                sell_wins += 1

            else:
                sell_loses += real_balance
                sell_misses += 1
                
            balance = sell_profit + sell_loses
        else:
            pass
            # print('######')
        df_row['real_balance'] = balance
        df_detailed_strategy_3 = df_detailed_strategy_3.append(df_row, ignore_index = True)
    
#     win_ratio = round(100*wins/(wins+misses), 2)
#     balance = round(profit + loses, 2)
        
#     data_resume['periodicity'].append(periodicity)
#     data_resume['profit'].append(profit)
#     data_resume['loses'].append(loses)
#     data_resume['wins'].append(wins)
#     data_resume['misses'].append(misses)
#     data_resume['win_ratio'].append(win_ratio)
#     data_resume['balance'].append(balance)
    
    buy_win_ratio = round(100*buy_wins/(buy_wins+buy_misses), 2)
    sell_win_ratio = round(100*sell_wins/(sell_wins+sell_misses), 2)
    wins = buy_wins + sell_wins
    misses = buy_misses + sell_misses
    total_win_ratio = round(100*wins/(wins+misses), 2)

    buy_balance = round(buy_profit + buy_loses, 2)
    sell_balance = round(sell_profit + sell_loses, 2)
    total_balance = round(buy_balance + sell_balance, 2)

    accuracy = (buy_wins+sell_misses)/(buy_wins+buy_misses+sell_misses+sell_wins)
    precision = buy_wins/(buy_wins+buy_misses)
    recall = buy_wins/(buy_wins+sell_misses)
    f1_score = 2*precision*recall/(precision+recall)
    
    data_resume['periodicity'].append(periodicity)
    data_resume['buy_profit'].append(buy_profit)
    data_resume['sell_profit'].append(sell_profit)
    data_resume['buy_loses'].append(buy_loses)
    data_resume['sell_loses'].append(sell_loses)
    data_resume['buy_wins'].append(buy_wins)
    data_resume['sell_wins'].append(sell_wins)
    data_resume['buy_misses'].append(buy_misses)
    data_resume['sell_misses'].append(sell_misses)
    data_resume['buy_win_ratio'].append(buy_win_ratio)
    data_resume['sell_win_ratio'].append(sell_win_ratio)
    data_resume['total_win_ratio'].append(total_win_ratio)
    data_resume['buy_balance'].append(buy_balance)
    data_resume['sell_balance'].append(sell_balance)
    data_resume['total_balance'].append(total_balance)
    data_resume['accuracy'].append(accuracy)
    data_resume['precision'].append(precision)
    data_resume['recall'].append(recall)
    data_resume['f1-score'].append(f1_score)

print("finish")
df_detailed_strategy_3 = df_detailed_strategy_3.sort_values(by = 'date_test').reset_index(drop = True)
df_resume_strategy_3 = pd.DataFrame(data_resume).sort_values(by = 'periodicity').reset_index(drop = True)
```
